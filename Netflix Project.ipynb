```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('netflix_titles.csv')

print(df.head())
print(df.info())

df['date_added'] = df['date_added'].str.strip()

df['date_added'] = pd.to_datetime(df['date_added'])

df['rating'].fillna('Not Rated', inplace=True)


print(df.isnull().sum())

# 5. Exploratory Data Analysis (EDA)

# a) Count of Movies vs TV Shows
plt.figure(figsize=(6,4))
sns.countplot(x='type', data=df)
plt.title('Count of Movies vs TV Shows')
plt.show()

# b) Top 10 Countries Producing Netflix Content
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.xticks(rotation=45)
plt.title('Top 10 Countries Producing Netflix Content')
plt.ylabel('Number of Titles')
plt.show()

# c) Number of Releases by Year
df['release_year'].value_counts().sort_index().plot(kind='line', figsize=(10,5))
plt.title('Number of Releases Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()

# 6. Final Insights
print("""
🔹 Netflix focuses more on Movies than TV Shows.
🔹 US and India are major contributors to Netflix's content library.
🔹 The number of new titles has grown steadily over time, peaking in recent years.
""")
```
